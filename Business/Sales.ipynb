```python
import pandas as pd

def consolidate_excel_files(states, output_file):
    all_data = pd.DataFrame()
    
    for state in states:
        file_name = f"{state}.xlsx"
        try:
            df = pd.read_excel(file_name)
            all_data = all_data.append(df, ignore_index=True)
        except FileNotFoundError:
            print(f"File '{file_name}' not found. Skipping...")

    all_data.to_excel(output_file, index=False)
    print("Consolidation completed. The data is saved to", output_file)

# Manually type each state here
states = ["MP", "Delhi", "Rajasthan", "Haryana", "Gujarat"]

# Set the output file to "indianstate.xlsx"
output_file = "indianstate.xlsx"

consolidate_excel_files(states, output_file)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_graph_from_excel(file_path, sheet_name):
    try:
        # Read the specific sheet from the Excel file
        df = pd.read_excel(file_path, sheet_name=sheet_name)

        # Assuming your data contains 'states', 'Google', 'No Website', and 'Website' columns
        states = df['States']
        google_visits = df['Google']
        no_website_visits = df['No Website']
        website_visits = df['Website']

        # Plotting the graph
        plt.figure(figsize=(12, 6))
        plt.bar(states, google_visits, label='Google', color='blue', width=0.5)
        plt.bar(states, no_website_visits, bottom=google_visits, label='No Website', color='orange', width=0.5)
        plt.bar(states, website_visits, bottom=google_visits + no_website_visits, label='Website', color='green', width=0.5)

        plt.xlabel('States')
        plt.ylabel('Number of Website Visits')
        plt.title('Electronic Sales by Indian States')
        plt.xticks(rotation=45, ha='right')
        plt.legend()
        plt.grid(True)
        plt.tight_layout()
        plt.show()

    except Exception as e:
        print("Error:", e)

# Replace the following path with the location of your 'indianstate.xlsx' file
file_path = "indianstate.xlsx"

# Replace 'Sheet3' with the name of the sheet you want to access and plot
sheet_name = "Sheet2"

plot_graph_from_excel(file_path, sheet_name)
```
