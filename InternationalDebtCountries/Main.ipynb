```python
import dash
from dash import dcc, html
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset and perform EDA (Assuming the dataset is in CSV format)
df = pd.read_csv('IDSData.csv')
```

```python
print(df)
```

```python
# Sort the DataFrame by total_debt in descending order
df_sorted = df.sort_values(by='2016', ascending=False)
```

```python
print(df_sorted)
```

```python
# EDA - Create a bar plot to show the total debt for each country
plt.figure(figsize=(12, 6))
sns.barplot(x='Country Name', y='2023', data=df_sorted)
plt.xticks(rotation=90)
plt.xlabel('Country')
plt.ylabel('Total Debt')
plt.title('Total Debt by Country')
plt.tight_layout()
plt.show()
```

```python
# Initialize the Dash app
app = dash.Dash("DEBT")

# Define the layout of the dashboard
app.layout = html.Div([
    dcc.Graph(
        id='debt-bar-plot',
        figure={
            'data': [
                {
                    'x': df_sorted['Country Name'],
                    'y': df_sorted['2023'],
                    'type': 'bar',
                    'name': 'Total Debt'
                }
            ],
            'layout': {
                'title': 'Total Debt by Country',
                'xaxis': {'title': 'Country'},
                'yaxis': {'title': 'Total Debt'}
            }
        }
    )
])

# Run the Dash app
if "DEBT" == "DEBT":
    app.run_server(debug=True)
```
