```python
import pandas as pd                          # pandas library is used for data manipulation and analysis
import numpy as np                           # numpy for numerical computing
import matplotlib.pyplot as plt              # matplotlib for plotting graphs and chart
import seaborn as sn
import warnings
warnings.filterwarnings(action='ignore')     # warning filter that ignores all warnings to avoid unnecessary messages or errors during data analysis.
```

```python
df1 = pd.read_csv("weather.csv")
df1.head()
```

```python
df1.shape
```

```python
df1.info()
```

```python
df1.describe()
```

```python
#To find any null values in given dataset
df1.isnull().sum()    
```

```python
#Dropping the null values from dataset for given attribute as low null values found

df1 = df1.dropna()
df1.isnull().sum()
```

```python
df1['RainTomorrow'].unique()
```

```python
Y = df1.RainTomorrow
Y.head()
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
df1['RainTomorrow']= label_encoder.fit_transform(df1['RainTomorrow'])
df1['RainTomorrow'].unique()
```

```python
label_encoder = preprocessing.LabelEncoder()
df1['WindGustDir']= label_encoder.fit_transform(df1['WindGustDir'])
df1['WindGustDir'].unique()
```

```python
label_encoder = preprocessing.LabelEncoder()
df1['RainToday']= label_encoder.fit_transform(df1['RainToday'])
df1['RainToday'].unique()
```

```python
#Correlation and Heatmap

hm = sn.heatmap(data = df1.corr(),annot=True,annot_kws={'size': 8})
# displaying the plotted heatmap
sn.set (rc = {'figure.figsize':(12, 12)})
plt.show()
```

```python
#from this heatmap we could make out that WindDir9am, WindDir3pm, WindSpeed9am is less correlated with Rain fall prediction 
X = df1.drop(['RainTomorrow','WindDir9am','WindDir3pm','WindSpeed9am'],axis='columns')
X.head()
```

```python
#Spliting the data in test and train

from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=10)
```

```python
#using SVM classifier with linear kernel

from sklearn import svm
clf = svm.SVC(kernel='linear') # Linear Kernel
clf.fit(X_train, Y_train)
y_pred = clf.predict(X_test)
```

```python
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(Y_test, y_pred))
```

```python
#using SVM classifier with Polynomial kernel

model = svm.SVC(kernel='poly')
model.fit(X_train, Y_train)
y_pred = model.predict(X_test)
```

```python
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(Y_test, y_pred))
```
